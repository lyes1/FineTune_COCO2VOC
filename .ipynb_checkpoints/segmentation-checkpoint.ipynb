# Spécification des chemins

```python
# Dossier pour sauvegarder les logs et le modèle entrainé 
MODEL_DIR = os.path.join(ROOT_DIR, "logs")

# Dossier des images sur lesquelles réaliser la detection
IMAGE_DIR = os.path.join(ROOT_DIR, "keras-mask-rcnn/images")

# Chemin relatif vers le fichier des labels des classes
CLASSES_LABELS_PATH = os.path.join(ROOT_DIR, "classes_labels.txt")

# Chemin relatif vers le fichier des points d'un modèle entrainé
PRE_TRAINED_MODEL_PATH = os.path.join(ROOT_DIR, "keras-mask-rcnn/mask_rcnn_coco.h5")

# Download COCO trained weights from Releases if needed
#if not os.path.exists(COCO_MODEL_PATH):
 #   utils.download_trained_weights(COCO_MODEL_PATH)
```

```python
# Charger les labels des classes depuis le fichier, un label par ligne (séparateur : "\n")
CLASS_NAMES = open(CLASSES_LABELS_PATH).read().strip().split("\n")
```

# Configuration

```python
class SimpleConfig(Config):
    # give the configuration a recognizable name
    NAME = "coco_inference"

    # set the number of GPUs to use along with the number of images
    # per GPU
    GPU_COUNT = 1
    IMAGES_PER_GPU = 1

    # number of classes (we would normally add +1 for the background
    # but the background class is *already* included in the class
    # names)
    NUM_CLASSES = len(CLASS_NAMES)
```

```python
# initialize the inference configuration
config = SimpleConfig()
config.display()
```

# Chargement du modèle

```python
# initialize the Mask R-CNN model for inference and then load the weights
model = modellib.MaskRCNN(mode="training", config=config, model_dir=os.getcwd())
model.load_weights(PRE_TRAINED_MODEL_PATH, by_name=True)
```

```python
# Charger une image aléatoire depuis le dossier images
file_names = next(os.walk(IMAGE_DIR))[2]
image = skimage.io.imread(os.path.join(IMAGE_DIR, random.choice(file_names)))

#image = skimage.io.imread ("/Users/lyes/Desktop/MD5/Untitled Folder/prediction/djamila.jpg")
```

```python
# Exécuter la detection (forward pass of the network to obtain the results)
results = model.detect([image], verbose=1)
```

```python
# Afficher les résultats
r = results[0]
visualize.display_instances(image, r['rois'], r['masks'], r['class_ids'], 
                            CLASS_NAMES, r['scores'])
```
