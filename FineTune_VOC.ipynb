# Finetuning

## Training Configuration

La classe VOC2012Config est déclarer dans un fichier séparé.<br>
NUM_CLASSES = 20 + 1  # + Background<br>
**Number of training steps per epoch:**<br>
STEPS_PER_EPOCH = 200<br>
**Validation steps:**<br>
VALIDATION_STEPS = 5<br>

```python
# VOC2012 Dataset
config_VOC2012 = VOC2012.VOC2012Config()
#tf.compat.v1.disable_eager_execution()

# Create model in training mode, saved in MODEL_DIR
model_voc = modellib.MaskRCNN(mode="training", config=config_VOC2012, model_dir=MODEL_DIR)
```

## COCO model's weights loading

```python
# Local path to COCO trained weights file
COCO_MODEL_PATH = os.path.join(ROOT_DIR, "mask_rcnn_coco.h5")
# Download COCO trained weights from Releases if needed
if not os.path.exists(COCO_MODEL_PATH):
    utils.download_trained_weights(COCO_MODEL_PATH)

# Set weights file path
# Use COCO trained weights file
coco_weights_path = COCO_MODEL_PATH

# Load weights
#skip layers that are different due to the different number of classes
print("Loading weights ", coco_weights_path)
model_voc.load_weights(coco_weights_path, by_name=True, 
                             exclude=[ "mrcnn_class_logits", "mrcnn_bbox_fc", 
                                      "mrcnn_bbox", "mrcnn_mask"])
```

```python
#tf.compat.v1.disable_eager_execution()
# Train the head branches
# Passing layers="heads" freezes all layers except the head
# layers. You can also pass a regular expression to select
# which layers to train by name pattern.
model_voc.train(voc2012_dataset_train, voc2012_dataset_val, 
            learning_rate=config_VOC2012.LEARNING_RATE, 
            epochs=2, 
            layers='heads')
```

# Detection

## Models configuration

### a. Our finetuned VOC model configuration

```python
# Create the model in inference mode for VOC 20 classification
model_voc_inf = modellib.MaskRCNN(mode="inference", config=config_VOC2012, model_dir=MODEL_DIR)

# Load the last trained model
# Set weights file path
last_model_weights_path = model_voc.find_last()
# Load VOC trained weights
print("Loading weights from ", last_model_weights_path)
model_voc_inf.load_weights(last_model_weights_path, by_name=True)
```

### b. COCO model configuration

```python
# MS COCO Dataset
config_COCO = coco.CocoConfig()
#config_COCO.display()

# Create model in inference mode for COCO 80 classification
model_coco = modellib.MaskRCNN(mode="inference", config=config_COCO, model_dir=MODEL_DIR)

# Load COCO weights
print("Loading weights ", coco_weights_path)
model_coco.load_weights(coco_weights_path, by_name=True)
```

## Image Loading

```python
# Charger une image aléatoire depuis le dossier images
#IMAGE_DIR = ROOT_DIR+"/../dataset_VOC2012/Test/VOC2012/JPEGImages/"
IMAGE_DIR = ROOT_DIR+"/../dataset_VOC2012/Train_val/VOC2012/JPEGImages/"

file_names = next(os.walk(IMAGE_DIR))[2]
image = skimage.io.imread(os.path.join(IMAGE_DIR, random.choice(file_names)))

display_images([image])
#image = skimage.io.imread ("/Users/lyes/Desktop/MD5/Untitled Folder/prediction/djamila.jpg")
```

## Inference using our finetuned VOC model 

```python
#Remettre a sa place
# Exécuter la detection (forward pass of the network to obtain the results)
results = model_voc_inf.detect([image], verbose=1)

# Afficher les résultats
r = results[0]
visualize.display_instances(image, r['rois'], r['masks'], r['class_ids'], 
                            voc2012_dataset_train.class_names, r['scores'], title="Predictions VOC")
```

```python
tupls = []
for i in range(0,len(r['class_ids'])):
    tupls.append((r['class_ids'][i],r['scores'][i]))
    
print("List of detected objects using the COCO model:")
for _, info in enumerate(voc2012_dataset_train.class_info):
    for tupl in tupls:
        if info['id']==tupl[0]:
            print("{:3}. {:10}:{:.3f}".format(tupl[0], info['name'], tupl[1]))
```

```python
# Exécuter la detection (forward pass of the network to obtain the results)
results = model_coco.detect([image], verbose=1)

# Afficher les résultats
r = results[0]
visualize.display_instances(image, r['rois'], r['masks'], r['class_ids'], 
                            coco_dataset.class_names, r['scores'], title="Predictions COCO")
```

```python
tupls = []
for i in range(0,len(r['class_ids'])):
    tupls.append((r['class_ids'][i],r['scores'][i]))
    
print("List of detected objects using the COCO model:")
for _, info in enumerate(coco_dataset.class_info):
    for tupl in tupls:
        if info['id']==tupl[0]:
            print("{:3}. {:10}:{:.3f}".format(tupl[0], info['name'], tupl[1]))
```
